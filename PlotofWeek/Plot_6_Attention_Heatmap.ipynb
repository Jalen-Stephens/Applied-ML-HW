```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import AutoTokenizer, AutoModel

print('torch:', torch.__version__)
```

## 1. Concepts (quick)

Self-attention computes, for each token, a weighted mixture of other token representations.

$$
\mathrm{Attention}(Q,K,V)=\mathrm{softmax}\left(\frac{QK^\top}{\sqrt{d}}\right)V
$$

The **attention heatmap** shows the softmax weights:
- Rows (queries): “where each token *looks* from”
- Columns (keys): “what each token *looks at*”


## 2. Load a model that returns attention

We'll use BERT (bidirectional attention) so we can visualize full token×token weights easily.


```python
model_name = 'bert-base-uncased'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name, output_attentions=True)
model.eval();

print('Loaded:', model_name)
```

## 3. Choose a sentence



```python
sentence = 'The cat sat on the mat because it was tired.'
inputs = tokenizer(sentence, return_tensors='pt')

with torch.no_grad():
    outputs = model(**inputs)

attentions = outputs.attentions  # tuple of length num_layers
tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])

num_layers = len(attentions)
num_heads = attentions[0].shape[1]
seq_len = attentions[0].shape[-1]

print('num_layers:', num_layers)
print('num_heads :', num_heads)
print('seq_len   :', seq_len)
print('tokens    :', tokens)
```

## 4. Plot a single attention head heatmap


```python
layer = 8   # try changing this
head  = 3   # try changing this

A = attentions[layer][0, head].detach().cpu().numpy()  # (seq_len, seq_len)

plt.figure(figsize=(9, 7))
sns.heatmap(A, xticklabels=tokens, yticklabels=tokens)
plt.xticks(rotation=90)
plt.title(f'Attention Heatmap — {model_name} — Layer {layer}, Head {head}')
plt.xlabel('Key tokens (attended-to)')
plt.ylabel('Query tokens (attending-from)')
plt.tight_layout()
plt.show()
```
